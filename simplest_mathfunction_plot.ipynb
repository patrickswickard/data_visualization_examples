```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

%reload_ext autoreload
%autoreload 2
```

```python
# for numpy arrays we can do a special shorthand "linspace" thing to get the x axis
# which is pretty useful for just plotting x vs f(x)
# taking a quick look under the hood shows how this works:
# this does 11 points between 0 and 5 (put in "one more" than you want since endpoints are inclusive)
x_1 = np.linspace(0,5,11)
y_1 = x_1**2
print(x_1)
print(y_1)
```

```python
plt.plot(x_1,y_1)
```

```python

# and we make a chunky plot of this
plt.title('CHART_TITLE')
plt.xlabel('X LABEL')
plt.ylabel('Y LABEL')
plt.plot(x_1,y_1)

# to display this outside of Jupyter:
# plt.show()
# to save this plot to a filename
# plt.savefig('image_from_jupyter.png')
```

```python
# one more with a finer plot and a trig function

x_2 = np.linspace(0,5*np.pi,100)
y_2 = np.sin(x_2)
print(x_2)
print(y_2)
plt.plot(x_2,y_2)
```

```python
# or even simpler:
x_2 = np.linspace(-2*np.pi,2*np.pi,100)
plt.plot(x_2,np.sin(x_2))
```
