```python
# so a big concept in making the syntax of matplotlib make sense
# is wrapping your head around the figures and axes
# a common mistake is to approach things in the sense of thinking things are
# packaged as individual graphs or plots
# first of all be aware that several commands work directly from plt
# which is the object we're playing with in the first place
# a figure is the overall package which often contains a single graph
# but may contain multiple graphs/plots side-by-side
# it is always an option to do one plot per figure
# but it is also an option not to
# the real unit you want to think of instead is axes
# a figure may have multiple axes, i.e. subplots
# which may be laid out side-by-side or in a grid
# we can ignore doing multiple axes for now and just know that
# the typical way (with a utility wrapper) of creating a single figure is
fig, ax = plt.subplots()
```

```python
# PARTS OF A FIGURE
# the whole plot thing plt which encompasses the general style, all figures you may have created, etc plt
# Figure - whole image which may have multiple "panels" called axes   plt.figure
# Axes - individual "panel" of a Figure  fig.subplots   or created with fig, ax = plt.subplots()
# Axes title - title for this axes panel   ax.set_title
# x axis - horizontal component of the axes   ax.xaxis
# y axis - vertical component of the axes ax.yaxis
# xlabel - label for x axis ax.set_xlabel
# ylabel - label for y axis ax.set_ylabel
# major tick (x axis) - the large ticks on the x axis  ax.xaxis.set_major_locator
# major tick (y axis) - the large ticks on the y axis  ax.yaxis.set_major_locator
# minor tick (x axis) - the small ticks on the x axis  ax.xaxis.set_minor_locator
# minor tick (y axis) - the small ticks on the y axis  ax.yaxis.set_minor_locator
# major tick label (x axis) - large x tick labels ax.xaxis.set_major_formatter
# major tick label (y axis) - large y tick labels ax.yaxis.set_major_formatter
# minor tick label (x axis) - small x tick labels ax.xaxis.set_minor_formatter
# minor tick label (y axis) - small y tick labels ax.yaxis.set_minor_formatter
# spines - the sides opposite labeled axis sides ax.spines
# grid - the grid inside the axes plot panel   ax.grid
# line - the line(s) making up the plot itself ax.plot
# markers - individual points marked in the axes  ax.scatter
# legend - a legend which can explain components  ax.legend
```

```python
# OTHER ESOTERIC TERMS
# Artist - everything visible in the figure is called an Artist (yes this is dumb and confusing)
# this includes Figure, Axes, Axis, Text, Line2D, collections (?), Patch (?), etc
# so if you see Artist in an error/diagnositic message it refers to some visible element
# Artists are displayed on the "canvas" to muddle this even farther, as if Van Gogh himself
# was splayed out upon a workspace
# most Artists are tied to a particular Axes and don't go leaping Axes to Axes or being shared
```

```python
# plotting functions typically expect input of type numpy.array or the moral equivalent
# and if you're not doing that then you're probably abusing something esoteric and may
# get surprising or disappointing behavior
# best practices is to explicitly convert to numpy.array before plotting
```

```python
# BEHOLD the glories of OO style (explicit):

x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(6.0, 9.0), layout='constrained')
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

```python
# implicit pyplot style lets Larry Wall read your mind and guess your intentions
# which is fantastic for old-school matlabperletc dongs like myself
# but leads to alcoholism as soon as you need to do something complicated
x = np.linspace(0, 2, 100)  # Sample data.

plt.figure(figsize=(6, 9), layout='constrained')
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```
