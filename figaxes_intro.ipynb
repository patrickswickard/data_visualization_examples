```python
# matplotlib actually has some good documentation...
# https://matplotlib.org/stable/users/explain/quick_start.html#quick-start

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

%reload_ext autoreload
%autoreload 2
```

```python
# matplotlib has implicit and explicit (object oriented) interfaces for plotting
# in general the explicit is more verbose but less confusing
# the figure is like a canvas, the axes are like a part of that 
# canvas upon which we make a particular visualization

x_1 = np.linspace(0,1,101)

fig, ax = plt.subplots()
ax.plot(x_1,x_1**2)
```

```python
# or here's a very simple horizontal bar chart

data = {'Barton LLC': 109438.50,
        'Frami, Hills and Schmidt': 103569.59,
        'Fritsch, Russel and Anderson': 112214.71,
        'Jerde-Hilpert': 112591.43,
        'Keeling LLC': 100934.30,
        'Koepp Ltd': 103660.54,
        'Kulas Inc': 137351.96,
        'Trantow-Barrows': 123381.38,
        'White-Trantow': 135841.99,
        'Will LLC': 104437.60}
group_data = list(data.values())
group_names = list(data.keys())

fig, ax = plt.subplots()
ax.barh(group_names,group_data)

```

```python
# so a big concept in making the syntax of matplotlib make sense
# is wrapping your head around the figures and axes
# a common mistake is to approach things in the sense of thinking things are
# packaged as individual graphs or plots
# a figure is the overall package which often contains a single graph
# but may contain multiple graphs/plots side-by-side
# it is always an option to do one plot per figure
# but it is also an option not to
# the real unit you want to think of instead is axes
# a figure may have multiple axes, i.e. subplots
# which may be laid out side-by-side or in a grid
# we can ignore doing multiple axes for now and just know that
# the typical way of creating a single figure is
fig, ax = plt.subplots()
```
